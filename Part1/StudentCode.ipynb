## Task # 3



```
Content Recommendation & Personalization

Big Picture: Understanding Data and Inclusion/Exclusion Logic
We'll use the 'merge' function for a left join, which is more explicit.

```

````python
# Merge 'titles_df' and 'credits_df' on the 'id' column.
merged_df = titles_df.merge(credits_df, on='id', how='left')


# Display the first few rows of the merged DataFrame.
merged_df.head()
````

````python
# Sample user input
user_wants_movie = True  # User wants to watch shows
user_wants_show = True    # User wants to watch shows
user_favorite_genre = "ENTER_A_GENRE"  # User's favorite genre
user_favorite_actor_director = "ENTER_AN_ACTOR/DIRECTOR"  # User's favorite actor/director , enter name txt
````

````python
# Initialize a set to store unique media recommendations, a place that will hold everything that gets recommended
media_recommendations = set()
````

````python
# Loop through the rows of the media data and make recommendations based on user input
for index, row in merged_df.iterrows():
    # Get information from the current row
    media_type = row["TYPEHERE"]  # Extract the media type (e.g., "MOVIE" or "SHOW")
    genres = row["TYPEHERE"]    # Extract the list of genres associated with the media
    actor_director = row["TYPEHERE"]  # Extract the actor or director's name

    # Check if the user wants to watch movies or shows
    if (
        (user_wants_movie and media_type == "TYPEHERE") or
        (user_wants_show and media_type == "TYPEHERE")
    ):
        # Check if the user's favorite genre matches any of the genres of the current media
        if user_favorite_genre in genres:
            # Check if the user has a favorite actor or director, the "" covers if they didn't put in a favorite actor/director
            if user_favorite_actor_director == "" or user_favorite_actor_director == actor_director:
                # If all conditions are met, add the media title and actor/director to recommendations
                media_recommendations.TYPEHERE((row["title"], actor_director))
````

````python
# Convert the set to a list for printing
media_recommendations_list = TYPEHERE(media_recommendations)
````

````python
# Check if there are any media recommendations in the 'media_recommendations_list'
if media_recommendations_list:  # This checks if the list is not empty
    print("Media Recommendations:")  # Display a message indicating media recommendations are being shown

    # Iterate through each recommendation and its associated actor/director
    for recommendation, actor_director in media_recommendations_list:
        # Print the title and the actor/director of each recommendation
        print(f"Title: {recommendation}, Actor/Director: {actor_director}")
else:
    # If the 'media_recommendations_list' is empty, display a message indicating no recommendations were found
    print("No media recommendations found based on your preferences.")
````
